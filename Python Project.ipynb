```python
import yfinance as yf

# Extract stock data for Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Save the data to a CSV file
tesla_data.to_csv('tesla_data.csv')

# Reset the index of the tesla_data dataframe
tesla_data = tesla_data.reset_index()

# Display the first five rows using the head() function
print(tesla_data.head())
```

```python
# Import the yfinance library
import yfinance as yf

# Set the ticker symbol and timeframe
ticker = "GME"
start_date = "2010-01-01"
end_date = "2021-12-31"

# Use yfinance to extract the stock data
gme_data = yf.download(ticker, start=start_date, end=end_date)

# Display the first five rows of the GME stock data
print(gme_data.head())
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
html_content = requests.get(url).text
```

```python
soup = BeautifulSoup(html_content, "html.parser")
```

```python
revenue_table = soup.find_all("table")[1]
rows = revenue_table.find_all("tr")
```

```python
data = []
for row in rows[1:]:
    cells = row.find_all("td")
    revenue_dict = {"date": cells[0].get_text(), "revenue": cells[1].get_text()}
    data.append(revenue_dict)
```

```python
gme_revenue = pd.DataFrame(data)
```

```python
print(gme_revenue.tail())
```
